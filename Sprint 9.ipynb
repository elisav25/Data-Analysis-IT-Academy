### 1. Calculadora del índice de masa corporal

Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos. Puedes obtener más información de su cálculo en: 🔗 Índice de masa corporal IMC que es y cómo se calcula.

La función debe clasificar el resultado en sus respectivas categorías.

Consejo: Intenta validar los datos previamente, para que envíe un mensaje de advertencia si los datos introducidos por el usuario/a no están en el formato adecuado o no toma valores razonables.

```python

def pedir_datos(mensaje, minimo=None, maximo=None):
    while True:
        try:
            valor = float(input(mensaje).strip().replace(",", "."))
            if valor is not None and valor < minimo:
                print(f"Tienes que introducir un valor mayor que {minimo}")
            elif valor is not None and valor > maximo:
                print(f"Tienes que introducir un valor menor que {maximo}")
            else:
                return valor 
        except ValueError:
            print("Tienes que introducir el valor en números!")


def calculo_imc(peso, altura):
    return peso / (altura ** 2)

def categorias(imc):
    if imc < 18.5:
        return "peso bajo"
    elif imc < 24.9:
        return "peso normal"
    elif imc < 29.9:
        return "sobrepeso"
    else:
        return "obesidad"
    
peso = pedir_datos("introduce tu peso en kg: ", 25, 250)
altura = pedir_datos("Introduce tu altura en metros: ", 1, 2.5)
imc = round(calculo_imc(peso, altura), 2)
resultado = categorias(imc)
```

```python
print(f"Tu IMC es {imc} y la categoria correspondente es {resultado}")
```

### 2. Conversor de temperaturas

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).

Consejo: Intenta validar los datos previamente, para que envíe un mensaje de advertencia si los datos introducidos por el usuario/a no están en el formato adecuado.

(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¿Lista? ¿Diccionario? ¿DataFrame?) en vez de escribir muchos if else en función de la temperatura de origen y la temperatura de destino.



```python
def ask_data():
    while True:
        temp_1 = input("Introduce the temperature unit you want to convert choosing among c for Celsius," \
                        " f for Fahrenheit, k for Kelvin and ra for Rankine: ").lower().strip()
        if temp_1 in ("c", "f", "k", "ra"):
            break
        else:
            print("Insert a valid option: c, f, k or ra!")

    while True:
        temp_2 = input("Introduce the temperature unit you want to convert to, choosing among c for Celsius," \
                        " f for Fahrenheit, k for Kelvin and ra for Rankine: ").lower().strip()
        if temp_2 in ("c", "f", "k", "ra"):
            break
        else:
            print("Insert a valid option: c, f, k or ra!")

    while True:
        try:
            value = float(input("Introduce the temperature value you want to convert: ").strip().replace(",", "."))
            break
        except ValueError:
            print("The value has to be written in numbers!")

    return temp_1, temp_2, value

names = {
    "c": "Celsius",
    "f": "Fahrenheit",
    "k": "Kelvin",
    "ra": "Rankine"
}

convertions = {
    ("c", "c"): lambda x: x,
    ("c", "f"): lambda x: x * 9/5 + 32,
    ("c", "k"): lambda x: x + 273.15,
    ("c", "ra"): lambda x: (x + 273.15)* 9/5,

    ("f", "f"): lambda x: x,
    ("f", "c"): lambda x: (x - 32) * 5/9,
    ("f", "k"): lambda x: (x - 32) * 5/9 + 273.15,
    ("f", "ra"): lambda x: x + 459.67,

    ("k", "k"): lambda x: x,
    ("k", "c"): lambda x: x - 273.15,
    ("k", "f"): lambda x: (x - 273.15) * 9/5 + 32,
    ("k", "ra"): lambda x: x * 9/5,

    ("ra", "ra"): lambda x: x,
    ("ra", "c"): lambda x: (x - 491.67) * 5/9,
    ("ra", "f"): lambda x: x - 459.67,
    ("ra", "k"): lambda x: x * 5/9
}


def converter(temp_1, temp_2, value):
    return convertions[(temp_1, temp_2)](value)

temp_1, temp_2, value = ask_data()
result = round(converter(temp_1, temp_2, value), 2)
```

```python
print(f"{value} {names[temp_1]} = {result} {names[temp_2]}")
```

### 3. Contador de palabras de un texto.

Escribe una función que dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las posibles casuísticas que hagan que el programa no funcione correctamente.

(EXTRA): ¿Cuál es la longitud media de las palabras del texto que has escrito? “¿Hola cómo va?” debería devolver (4+3+2) / 3 = 3

```python
texto = "Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam," \
        " quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore" \
        " eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum."

def limpiar_palabras(texto):
    for signo in ",.;:?¿!¿'":
        texto = texto.replace(signo, "")
    return texto 

def contar_palabras(texto):
    palabras = texto.lower().split()
    elenco = {}
    for palabra in palabras:
        if palabra not in elenco:
            elenco[palabra] = 1
        else:
            elenco[palabra] += 1
    return elenco 

def longitud_media(texto):
    palabras = texto.lower().split()
    contador = 0
    for palabra in palabras:
        contador += len(palabra)
    return contador / len(palabras)

texto = limpiar_palabras(texto)
resultado = contar_palabras(texto)
longitud = longitud_media(texto)
```

```python
print(resultado)
print(f"La longitud media de las palabras es de {longitud:.2f} letras")
```

### 4. Diccionario inverso (con posibilidad de duplicados)

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.

```python
diccionario_original = {"a": 1, "b": 2, "c": 3, "d": 1}

def agrupar_claves(diccionario):
    diccionario_inverso = {}
    for clave, valor in diccionario.items():
        if valor not in diccionario_inverso:
            diccionario_inverso[valor] = [clave]
        else:
            diccionario_inverso[valor].append(clave)
    return diccionario_inverso
resultado = agrupar_claves(diccionario_original)

def mostrar_duplicados(diccionario_inverso):
    for valor, claves in diccionario_inverso.items():
        if len(claves) > 1:
            print(f"Valor duplicado: {valor}")
            print(f"Claves asociadas: {claves}")
duplicados = mostrar_duplicados(resultado)
```

### 1. Contador y ordenador de palabras de un texto.

El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z. "tu_me_quieres_blanca.txt" la salida esperada sería:

```python
ruta_archivo = r"C:\Users\elisa\Documents\Data Analytics\Sprint 9\tu me quieres blanca.txt"

import string

def limpiar_texto(ruta_archivo):
    with open(ruta_archivo, "r", encoding="utf-8") as f:
        texto = f.read().lower()

    for simbolo in string.punctuation:
        texto = texto.replace(simbolo, "")

    return texto

def contar_palabras(texto):
    diccionario_palabras = {}
    for palabra in texto.split():
        if palabra not in diccionario_palabras:
            diccionario_palabras[palabra] = 1
        else:
            diccionario_palabras[palabra] += 1
    return diccionario_palabras

def agrupar_ordenar(diccionario_palabras):
    diccionario_final = {}
    for palabra, cuenta in diccionario_palabras.items():
        inicial = palabra[0]
        if inicial not in diccionario_final:
            diccionario_final[inicial] = {}
        diccionario_final[inicial][palabra] = cuenta

    diccionario_final_ordenado = {}
    for letra in sorted(diccionario_final.keys()):
        diccionario_final_ordenado[letra] = dict(sorted(diccionario_final[letra].items()))
    return diccionario_final_ordenado

texto = limpiar_texto(ruta_archivo)
dic_palabras = contar_palabras(texto)
resultado = agrupar_ordenar(dic_palabras)
```

```python
for letra, subdict in resultado.items():
    print(f"{letra} : {subdict}\n")
```

### 2. Conversión de tipos de datos.

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente:

[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
datos = ['1.3', 'one', '1e10', 'seven', '3-1/2',
         ('2', 1, 1.4, 'not-a-number'),
         [1, 2, '3', '3.4']]

convertibles = []
no_convertibles = []

def procesar(elemento):
    if isinstance(elemento, (tuple, list)):
        for item in elemento:
            procesar(item)
    else:
        try:
            num = float(elemento)
            convertibles.append(num)
        except:
            no_convertibles.append(elemento)

procesar(datos)
```

```python
print(f"Elementos convertibles: {convertibles}")
print(f"Elementos no convertibles: {no_convertibles}")
```
